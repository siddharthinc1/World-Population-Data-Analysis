# Data Visualization

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("D:\Data Analyst\Projects\Analysis of World Population Data\world_population_data1.csv")
df1
```

```python
Pop2023 = df1.groupby(['Continent'])[['2023 population','2022 population','2020 population','2015 population','2010 population','2000 population','1990 population','1980 population','1970 population']].sum().sort_values(by=['2023 population'],ascending=False).reset_index()
Pop2023
```

```python
Pop2023.plot(kind='bar',x='Continent', y=['2023 population','2022 population','2020 population','2015 population','2010 population','2000 population','1990 population','1980 population','1970 population'])
plt.title('World population growth in Year 2023,2022,2020,2015,2010,2000,1990,1980,1970')
```

```python
C_area = df1.groupby('Continent')['Area (km²)'].sum().reset_index()
C_area
```

```python
C_area.plot(kind='pie', autopct="%.0f%%")
plt.title('Continents with its area (km²)')
plt.legend()
plt.show()
```

```python
c_country = df1.groupby('Continent')['Country'].count().reset_index()
c_country
```

```python
c_country.plot(kind='bar', x='Continent',y='Country')
plt.title("Number of countries in each continents")
```

```python
Top_c = df1[['Continent','Country','2023 population']].sort_values(['2023 population'],ascending=False, ignore_index=True).iloc[:10]
Top_c
```

```python
Top_c.plot(kind='bar', x='Country', y=['2023 population']) 
plt.title('Top 10 high population countries in the world in Year 2023')
```

```python
Low_c = df1[['Country','2023 population','Area (km²)']].sort_values(by=['2023 population'],ignore_index=True).iloc[:5]
Low_c
```

```python
Low_c.plot(kind='bar', x='Country',y=['2023 population','Area (km²)'])
plt.title('Top 5 low population countries in the world in Year 2023')
```

```python
Areal = df1[['Country', 'Area (km²)','2023 population']].sort_values('Area (km²)',ascending=False,ignore_index=True).head(5)
Areal
```

```python
Areal.plot(kind='bar',x='Country',y=['Area (km²)'])
plt.title('Top 5 countries which have large area(km²)')
```

```python
density = df1[['Country','Density (km²)','2023 population','Area (km²)']].sort_values(['Density (km²)'],ignore_index=True).head(10)
density
```

```python
density.plot(kind='bar', x='Country', y='Density (km²)')
plt.title('Top 10 low population density (km²) countries')
```
